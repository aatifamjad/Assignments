Q1. What are the three measures of central tendency?

Answer 1)

The three measures of central tendency are mean, median, and mode.

Q2. What is the difference between the mean, median, and mode? How are they used to measure the
central tendency of a dataset?

Answer 2)

Mean is the sum of all values in a dataset divided by the total number of values. Median is the middle value in an ordered dataset. Mode is the value that appears most frequently in a dataset. These measures are used to provide a central value that represents the typical value of the dataset.

Q5. How are measures of dispersion such as range, variance, and standard deviation used to describe
the spread of a dataset? Provide an example.

Answer 5)

Measures of dispersion such as range, variance, and standard deviation are used to describe the spread or variability of a dataset. The range is the difference between the highest and lowest values. The variance is the average of the squared differences between each value and the mean. The standard deviation is the square root of the variance. These measures help to provide a more complete understanding of the dataset beyond the central tendency measures.

Q8. What do you understand about skewness in data?

Answer 8)

Skewness in data refers to the degree of asymmetry of the dataset. If the dataset is skewed, it means that it does not have a symmetrical distribution, and one tail may be longer than the other. Skewness can affect the measures of central tendency and dispersion.


Q9. If a data is right skewed then what will be the position of median with respect to mean?

Answer 9)

If a dataset is right skewed, the median will be less than the mean.


Q10. Explain the difference between covariance and correlation. How are these measures used in
statistical analysis?


Answer 10)


 Covariance measures the direction and strength of the linear relationship between two variables. Correlation measures the degree to which two variables are related to each other. Correlation takes the values between -1 and 1, whereas covariance can take any value. These measures are used to identify the relationship between two variables and to make predictions.
 
 
 Q11. What is the formula for calculating the sample mean? Provide an example calculation for a
dataset.

Answer 11) The formula for calculating the sample mean is:

mean = (sum of values) / (number of values)

For example, if a dataset has values 2, 4, 6, and 8, the mean would be (2 + 4 + 6 + 8) / 4 = 5.

Q13. How is covariance different from correlation?


Answer 13)

Covariance measures the direction and strength of the linear relationship between two variables. Correlation measures the degree to which two variables are related to each other, but it also standardizes the measure to be between -1 and 1, making it easier to interpret and compare.
