Q8: What is the z score? State the importance of the z score.

Answer 8)

The z score, also known as the standard score, is a statistical measure that indicates how many standard deviations an observation or data point is above or below the mean of the data set. It is calculated as the difference between the observed value and the mean of the population, divided by the standard deviation of the population. The z score is important because it helps in standardizing different variables with different units of measurement, making it easier to compare them. It is also used to determine the probability of an observation occurring within a given range of values.


Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.

Answer 9)

The Central Limit Theorem (CLT) is a statistical theory that states that as the sample size increases, the distribution of sample means approaches a normal distribution, regardless of the shape of the original population distribution. This means that if we take a large number of random samples from a population, the means of those samples will tend to cluster around the population mean, with a standard deviation equal to the standard error of the mean. The significance of the CLT is that it provides a basis for making statistical inferences about a population from a sample, even when the population distribution is unknown or not normal.


Q10: State the assumptions of the Central Limit Theorem.


Answer 10)

The Central Limit Theorem makes certain assumptions about the sample and the population. These assumptions include:

a) Random sampling: The sample is randomly selected from the population, so that each individual in the population has an equal chance of being selected.

b) Independent observations: Each observation in the sample is independent of the others, meaning that the value of one observation does not affect the value of another.

c) Finite population or infinite population with replacement: The population from which the sample is drawn is either finite or infinite, but if it is infinite, the sample is drawn with replacement.

d) Sample size: The sample size is sufficiently large (usually n ≥ 30) to ensure that the sample means are normally distributed.
