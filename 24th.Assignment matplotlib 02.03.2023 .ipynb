Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Answer 1....

Matplotlib is a Python library used for data visualization. It provides a wide range of tools for creating various types of plots and graphs, making it a popular choice for scientists, engineers, and data analysts.

Matplotlib is used for creating plots and visualizations that help in understanding data patterns, trends, and relationships. It can be used to generate visualizations for various types of data, including 2D and 3D plots, histograms, scatter plots, bar charts, and pie charts, among others.

Pyplot is a sub-module of Matplotlib that provides a simple interface for creating and customizing plots. Some of the most common plots that can be created using Pyplot module are:

1) Line plots: These plots are used to represent data points on a continuous line, making it easy to visualize trends and patterns in the data.

2) Scatter plots: These plots use dots to represent data points, with the position of the dots indicating the values of the variables being plotted. Scatter plots are useful for identifying trends, clusters, and outliers in the data.

3) Bar charts: These plots are used to represent categorical data as rectangular bars, with the height of the bars indicating the values of the variables being plotted. Bar charts are useful for comparing the values of different categories.

4) Histograms: These plots are used to represent the distribution of a continuous variable, by dividing the data into bins and plotting the frequency of data points in each bin.

5) Pie charts: These plots are used to represent categorical data as slices of a circular pie, with the size of each slice indicating the proportion of the data in each category. Pie charts are useful for visualizing the relative proportions of different categories.


Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

Answer 2....


A scatter plot is a type of plot that displays values for two variables as a collection of points. It shows how much one variable is affected by another variable. Each point in a scatter plot represents the values of two variables. It is commonly used to observe relationships between variables and to identify trends or patterns.

To generate a scatter plot using the given data in Python, we can use the Matplotlib library.

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.title("Scatter Plot of X and Y")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

Answer 3 ...

The subplot() function is used to create multiple plots in a single figure. It takes three arguments, nrows, ncols, and index, where nrows and ncols are used to specify the number of rows and columns of subplots, and index is used to specify the index of the current subplot.



```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data for the four lines
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure with four subplots arranged in a 2x2 grid
fig, axs = plt.subplots(nrows=2, ncols=2)

# Plot the four lines in the four subplots
axs[0, 0].plot(x1, y1)
axs[0, 1].plot(x2, y2)
axs[1, 0].plot(x3, y3)
axs[1, 1].plot(x4, y4)

# Set the titles for the subplots
axs[0, 0].set_title('Line 1')
axs[0, 1].set_title('Line 2')
axs[1, 0].set_title('Line 3')
axs[1, 1].set_title('Line 4')

# Set the overall title for the figure
fig.suptitle('Four Line Plots')

# Display the figure
plt.show()
```

Answer 4....

A bar plot is a type of graph that represents data using rectangular bars. The height or length of the bar is proportional to the value being represented. Bar plots are used to compare different categories of data or to track changes over time. They are commonly used in business and economics to represent financial data, such as sales or profits.

```python
#  example of how to plot a bar plot and a horizontal bar plot
import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar plot
plt.bar(company, profit)
plt.title('Company Profits')
plt.xlabel('Company')
plt.ylabel('Profit ($)')
plt.show()

# Horizontal bar plot
plt.barh(company, profit)
plt.title('Company Profits')
plt.xlabel('Profit ($)')
plt.ylabel('Company')
plt.show()
```

```python
# example of how to plot a box plot  
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

plt.boxplot(data)
plt.title('Box Plot')
plt.ylabel('Value')
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.show()
```
